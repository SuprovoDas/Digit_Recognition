```python
import os
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import cv2

# Set parameters
output_dir = "synthetic_digits"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
font_size = 250
image_size = (256,256)  # Adjust size as needed
num_samples_per_digit = 100  # Number of images per digit

# Load a font (you can download different fonts or use system fonts)
font_path = "Lato-Bold.ttf"  # Replace with the path to a TTF font on your system
font = ImageFont.truetype(font_path, font_size)

```

```python
# Function to generate a single synthetic image of a digit
def generate_digit_image(digit, font, image_size, rotation=0, noise_level=0.1):
    # Create blank white canvas
    image = Image.new('L', image_size, color=255)
    draw = ImageDraw.Draw(image)
    
    # Calculate text position
    text = str(digit)
    text_size = draw.textsize(text, font=font)
    text_position = ((image_size[0] - text_size[0]) // 2, (image_size[1] - text_size[1]) // 2)
    
    # Draw digit in black
    draw.text(text_position, text, font=font, fill=0)
    
    # Rotate image
    if rotation != 0:
        image = image.rotate(rotation, expand=1, fillcolor=255)
    
    # Convert to numpy array for noise application
    image_np = np.array(image)
    
    # Apply random noise
    noise = np.random.normal(0, noise_level * 255, image_np.shape)
    noisy_image = image_np + noise
    noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)
    
    return noisy_image

```

```python

# Generate images for each digit
for digit in range(10):
    digit_dir = os.path.join(output_dir, str(digit))
    os.makedirs(digit_dir, exist_ok=True)
    
    for i in range(num_samples_per_digit):
        # Vary rotation and noise level
        rotation = np.random.randint(-30, 30)  # Random rotation angle
        noise_level = np.random.uniform(0.1, 0.4)  # Random noise level
        
        # Generate image
        image = generate_digit_image(digit, font, image_size, rotation, noise_level)
        
        # Save image
        filename = os.path.join(digit_dir, f"{digit}_{i}.png")
        cv2.imwrite(filename, image)

print("Synthetic dataset generated successfully!")
```
